```python
#Implementation of Two Player Tic-Tac-Toe game in Python.

''' We will make the board using dictionary 
    in which keys will be the location(i.e : top-left,mid-right,etc.)
    and initialliy it's values will be empty space and then after every move 
    we will change the value according to player's choice of move. '''

theBoard = {'1':' ' , '2':' ' , '3':' ' ,
            '4':' ' , '5':' ' , '6':' ' ,
            '7':' ' , '8':' ' , '9':' ' }

board_keys = []                         #inputs by user

for key in theBoard:
    board_keys.append(key)


''' We will have to print the updated board after every move in the game and 
    thus we will make a function in which we'll define the printBoard function
    so that we can easily print the board everytime by calling this function. '''

def PrintBoard(board):
    print(board['1'] + '|' + board['2'] + '|' + board['3'] )
    print("-+-+-")
    print(board['4'] + '|' + board['5'] + '|' + board['6']  )
    print("-+-+-")
    print(board['7'] + '|' + board['8'] + '|' + board['9']  )

#Now we will write the main function which will have all the funtionality
def game():

    turn = 'X'
    count = 0


    for i in range(10):
        PrintBoard(theBoard)
        print("Its your turn " + turn + ".Move to which place? ")

        move = input()

        if theBoard[move]==' ':
            theBoard[move]=turn
            count+=1
        else:
            print("That place is already filled.\nMove to which place?")
            continue

          # Now we will check if player X or O has won,for every move after 5 moves.
        if count>=5:
              if theBoard['1'] == theBoard['2'] == theBoard['3'] != ' ':                             #horizontal-upper 
                  PrintBoard(theBoard)
                  print("\nGame Over.\n")
                  print(" ***** "+ turn + "won. ***** ")
                  break
              elif theBoard['4']==theBoard['5']==theBoard['6']!= ' ':                           #horizontal-middle 
                  PrintBoard(theBoard)
                  print("\nGame Over.\n")
                  print(" ***** "+ turn + "won. ***** ")
                  break
              elif theBoard['7']==theBoard['8']==theBoard['9']!= ' ':                           #horizontal-lower 
                  PrintBoard(theBoard)
                  print("\nGame Over.\n")
                  print(" ***** "+ turn + "won. ***** ")
                  break
              elif theBoard['1']==theBoard['4']==theBoard['7']!= ' ':                           #vertical-leftmost 
                  PrintBoard(theBoard)
                  print("\nGame Over.\n")
                  print(" ***** "+ turn + "won. ***** ")
                  break
              elif theBoard['2']==theBoard['5']==theBoard['8']!= ' ':                           #vertical-middle 
                  PrintBoard(theBoard)
                  print("\nGame Over.\n")
                  print(" ***** "+ turn + "won. ***** ")
                  break
              elif theBoard['3']==theBoard['6']==theBoard['9']!= ' ':                           #vertical-rightmost 
                  PrintBoard(theBoard)
                  print("\nGame Over.\n")
                  print(" ***** "+ turn + "won. ***** ")
                  break
              elif theBoard['1']==theBoard['5']==theBoard['9']!= ' ':                           #diagonal-right
                  PrintBoard(theBoard)
                  print("\nGame Over.\n")
                  print(" ***** "+ turn + "won. ***** ")
                  break
              elif theBoard['3']==theBoard['5']==theBoard['7']!= ' ':                           #diagonal-left
                  PrintBoard(theBoard)
                  print("\nGame Over.\n")
                  print(" ***** "+ turn + "won. ***** ")
                  break

        #If neither X or O wins and the board is full we'll declare the result as "tie"
        if count==9:
            print("Game Over.\n")
            print("It's a Tie!!")

        #Now to change player after every turn
        if turn=='X':
          turn='O'
        else:
            turn='X'

    #Now we'll ask players if they want to restart the game or not
    restart = input("Restart the game?(y/n)")
    if restart =="y" or restart =="Y":
        for key in board_keys:
            theBoard[key]=' '

        game()

if __name__ == "__main__":
    game()
```
